## 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed', linewidth='2')
plt.plot(x, y_2,
         label='Y2 values', c='blue', lw='1.5')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) #opacity of gridlines
plt.legend(loc='upper left')
plt.savefig('cask.jpg', dpi=150) #dpi is to set resolution
plt.show()
```

### Things to note

| **Long form**   | **Abbreviation**                            |
|--------------------|--------------------------------------|
|color	| c |
| linestyle |	ls |
| linewidth	| lw |

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = 0.5, 2

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='purple', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, fmt='-', ecolor='red', capsize=2, elinewidth=1,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## We need another axis!

```python
x = np.linspace(0, 6, 100) ## generates an array x of 100 evenly spaced values starting from 0 and ending at 6
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
# plt.style.use('fivethirtyeight')
# plt.xkcd()
plt.style.use('default')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])


plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
