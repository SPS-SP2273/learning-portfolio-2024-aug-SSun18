```python
import os, glob, shutil
```

## Task 1

```python

calendar = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

sgoctfile = 'oct-08_singapore.txt'

def conv_mth(file):
    
    for mth in calendar.items():
        file = file.replace(mth[0], mth[1], 1)
        file = file.replace("_", "-")
    return file

conv_mth(sgoctfile)
```

## Task 2

```python
def rename_my_file(file):
    for mth in calendar:
        file = file.replace(mth, calendar[mth], 1)
        file = file.replace("_", "-")
    return file
    
rename_my_file(sgoctfile)
```

## Task 3

```python
file_names = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

new_file_names = file_names.copy()

index = 0
for file in file_names:
    new_file_names[index] = rename_my_file(file)
    print(f'{file_names[index]} ----> {new_file_names[index]}')
    index =+ 1
```

## Task 4

```python

# Use '/' or '\\' to subdivide folders as just '\' is taken as an escape sequence.

for file in glob.glob('os-collaboration-exercise-data\\*.txt', recursive = True):
    new_file = rename_my_file(file)
    shutil.move(file, new_file)
#    print(f'Moved logo to {new_file}')



glob.glob('os-collaboration-exercise-data\\**.txt', recursive = True)
```

## Task 5

```python
countries = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for country in countries:
    country_path = os.path.join('os-collaboration-exercise-data', country)
    try:
        os.mkdir(country_path)
        print(f'Creating {country_path}')
    except FileExistsError:
        print(f'{country_path} already exists; skipping creation.')
```
